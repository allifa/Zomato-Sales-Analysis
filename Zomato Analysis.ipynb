```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

```python
#we remove the rating as it will be a future hinderance
def handleRate(value):
    value = str(value).split("/")
    value = value[0];
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())   
```

```python
dataframe.info()
```

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'],  hue=dataframe['listed_in(type)']) 
plt.xlabel("type of restaurant")
plt.show()
```

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
#votes is in [] because pd requires us to use [] for the coloum to be added.
# double brackets would be used if we wanted to add 2 colomns and so on.
result = pd.DataFrame({'votes' : grouped_data})
plt.plot(result, c = "green", marker = "o")
plt.xlabel("type of restaurant", c = "blue", size = 20)
plt.ylabel("votes", c = "red", size = 20)
```

```python
plt.hist(dataframe['rate'], bins = 10)
plt.title("ratings distribution")
plt.show()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)
```

```python
pivot_table = dataframe.pivot_table(index = 'listed_in(type)', columns = 'online_order', aggfunc = 'size', fill_value=0)
'''If you have multiple records with the same
listed_in(type) and online_order values, 
using aggfunc='size' will 
simply count how many record
there are for each combination'''
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt = 'd')
plt.title("Heatmap")
plt.xlabel("online order")
plt.ylabel("listed in (type)")
plt.show()
```
